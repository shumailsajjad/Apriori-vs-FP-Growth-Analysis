```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
#loading market basket dataset..

df = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
```

```python
df.head()
```

```python
## Data Cleaning step

# replacing empty value with 0.
df.fillna(0,inplace=True)
```

```python
# Data Pre-processing step

# for using aprori , need to convert data in list format..
# transaction = [['apple','almonds'],['apple'],['banana','apple']]....

transactions = []

for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
## verifying - by printing the 1st transaction
transactions[1]
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df.head()
```

```python
frequent_itemsets_fp=fpgrowth(df, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets_fp
```

```python
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.3)
```
